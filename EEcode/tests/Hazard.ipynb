## Analysis

```
# Define our area of interest and date of change
aoi = ee.Geometry.Polygon(
        [[[-83.37017153264765, 37.48081395204879],
          [-83.37486536622822, 37.31933288374584],
          [-83.05319468739128, 37.30974497135589],
          [-83.05556035288436, 37.47934591201635]]])

doi = ee.Date('2018-08-01')
```

```
dictionary = dictionaries.forest
#print(dictionary.getInfo())
output = analyze.analyze_iw(aoi, doi, dictionary, 0)
```

```
# View the outputs using Leaflet
vis_params = {
        'palette': ['ffffff', 'd63000'],
        'min': 80,
        'max': 200}
        
my_map = folium.Map(location=[37.4, -83.15], zoom_start=12)

# Add the elevation model to the map object.
my_map.add_ee_layer(output.get('chi2'), vis_params, 'chi2')

# Add a layer control panel to the map.
my_map.add_child(folium.LayerControl())
```
